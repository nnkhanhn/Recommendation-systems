```python
from data_loader import dataloader
import matplotlib.pyplot as plt
user,ratings,movies = dataloader()
```

```python
import pandas as pd
final_df = ratings.merge(movies, on='movie_id', how='left').merge(user, on='user_id', how='left')
final_df.head()
```

```python
final_df_matrix = final_df.pivot(index='user_id',
                                 columns='movie_id',
                                 values='rating').fillna(0)
final_df_matrix.head()
```

```python
user_profile = pd.read_csv("u_dict.csv")
print(user_profile.head())
```

```python
profile = final_df_matrix.merge(user_profile, on="user_id", how="left")
profile.head()
```
