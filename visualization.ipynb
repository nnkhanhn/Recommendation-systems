```python
from data_loader import dataloader
import matplotlib.pyplot as plt
user,ratings,movies = dataloader()
print(user.info())
print(movies.head())
print(ratings.head())
```

```python
tags = [i.split("|") for i in movies["genres"].unique()]
columns = list(set([i for lst in tags for i in lst]))
print(columns)
for col in columns:
    movies[col] = movies["genres"].apply(lambda x: 1 if col in x else 0)
num_genres = []
for col in columns:
    for i in movies[col]:
        z = movies[col].sum()
    num_genres.append(z)
plt.bar(columns, num_genres)
plt.show()
```

```python
rt = [1,2,3,4,5]
k = []
for j in rt:
    z= 0
    for i in ratings['rating']:
        if i==j : z+=1
    k.append(z)

plt.bar(rt, k)
plt.show()
```
