```python
from data_loader import dataloader
import matplotlib.pyplot as plt
user,ratings,movies = dataloader()
print(user.head())
print(movies.head())
print(ratings.head())
```

```python
tags = [i.split("|") for i in movies["genres"].unique()]
columns = list(set([i for lst in tags for i in lst]))
print(columns)
for col in columns:
    movies[col] = movies["genres"].apply(lambda x: 1 if col in x else 0)
num_genres = []
for col in columns:
    for i in movies[col]:
        z = movies[col].sum()
    num_genres.append(z)
plt.bar(columns, num_genres)
plt.show()
```

```python
rt = [1,2,3,4,5]
k = []
for j in rt:
    z= 0
    for i in ratings['rating']:
        if i==j : z+=1
    k.append(z)

plt.bar(rt, k)
plt.show()
```

```python
import numpy as np
l = []
m = []
for i in ratings["movie_id"]:
    l.append(i)
for i in movies["movie_id"]:
    m.append(l.count(i))
    
print(m)

        
```

```python
import numpy as np 
print(np.mean(m))
print(m.count(0))

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ["not_rated","rated"]
k = m.__len__()-m.count(0)
pie = [m.count(0),k]
ax.pie(pie, labels = langs,autopct='%1.2f%%')
ax2 = fig.add_axes([0,1,1,1])
ax2.axis('equal')
langs2 = ('rated', 'sum')
pie2 = [np.sum(m),6040*3883-np.sum(m)]
ax2.pie(pie2, labels= langs2,autopct='%1.2f%%' )
plt.show()
```
